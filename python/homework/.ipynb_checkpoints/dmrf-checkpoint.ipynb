```python
import requests
import pandas as pd
from tqdm.auto import tqdm
from pandas.io.json import json_normalize
```

## Список объектов

```python
offset_ = 2001
limit_ = 1000
url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/kn/object?offset={offset_}&limit={limit_}&sortField=devId.devShortCleanNm&sortType=asc&objStatus=0'
res = requests.get(url)
objects_data = res.json()
```

```python
objects_data.keys()
```

```python
objects_list = objects_data.get('data').get('list')
```

```python
objids = [x.get('objId') for x in objects_list]
```

# Данные по определенному объекту

### Собрать информацию о всех строящихся объектах на сайте "наш.дом.рф"
### Cохранить ее в pandas dataframe

```python
df = pd.DataFrame()
for i in tqdm(range(len(objids))):
    url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/object/{objids[i]}'
    try:
        res = requests.get(url)
        df_new = json_normalize(res.json()['data'])
    except Exception as e:
        print(e)
        continue
    finally:
        if len(df) > 0:
            df = pd.concat([df, df_new], ignore_index=True)
        else:
            df = df_new
```

### Сравните среднюю цену квадратного метра в городах этого региона

#### objPriceAvg

```python
#df[['developer.regRegionDesc', 'developer.devOrgRegRegionCd']]
#df[['developer.regRegionDesc', 'developer.devOrgRegRegionCd']].set_index('developer.devOrgRegRegionCd').to_dict()
```

```python
df_1 = df[['region', 'address', 'objPriceAvg']].copy()
df_1['city'] = df['address'].str.split(',')
df_1['city'] = df_1['city'].str[0]
df_1['city'] = df_1['city'].str.split().apply(lambda x: ''.join(list(filter(str.istitle, x))))
```

```python
df_2 = df_1.groupby(['region', 'city']).mean()
df_2 = df_2[df_2.objPriceAvg.notnull()]
df_2
```

### Определите количество строящихся объектов и квартир в регионах

```python
df_3 = df[['photoRenderDTO', 'region', 'objSquareLiving']]
df_3['photoRenderDTO'] = df_3[df_3['photoRenderDTO'].notnull()]
df_3 = df_3[df_3['photoRenderDTO'].apply(lambda x: len(x)) > 0].reset_index()
df_3['objReadyDesc'] = [df_3['photoRenderDTO'].str[0][i]['objReadyDesc'] for i in range(len(df_3))]
df_3 = df_3[df_3.objReadyDesc == 'Строится']
df_3[['region','objReadyDesc']].groupby(['region']).count()
```

### Определите объем строящегося жилья в квадратных метрах

```python
df_3['objBuildTypeShortDesc'] = [df_3['photoRenderDTO'].str[0][i]['objBuildTypeShortDesc'] for i in range(len(df_3))]
df_4 = df_3[df_3.objBuildTypeShortDesc.str.lower() == 'жилое']
df_4[['region','objSquareLiving']].groupby('region').sum()
```
