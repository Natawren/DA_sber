```python
df = pd.read_sql("""
                    select * from distribution_by_type
                """, conn)
df
```

```python
df_a = df[df.maker == 'A']
fig1, ax1 = plt.subplots()
ax1.pie(df_a.proc, labels=df_a.type, autopct='%1.1f%%')
ax1.axis('equal')  
plt.show()
```
